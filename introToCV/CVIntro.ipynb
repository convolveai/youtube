```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Read Image

```python
image = cv2.imread('lena_std.tif')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
```

```python
image
```

## Colored to Grayscale

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap='gray')
```

```python
image_gray.shape
```

## Edge Detection

```python
edges = cv2.Canny(image_gray,100,200)
 
plt.subplot(121),plt.imshow(image_gray,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
```

## Convolutions and Related Applications (Blurring)

```python
blur = cv2.blur(image,(5,5))
 
plt.subplot(121),plt.imshow(image),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

## Convolutions form basis for Convolutional Neural Networks - Convolution is the basic operation in CNNs
